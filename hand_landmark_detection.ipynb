```python
import cv2
import mediapipe as mp
```

```python
mphands=mp.solutions.hands
mp_drawing=mp.solutions.drawing_utils
```

```python
hand=mphands.Hands()
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=hand.process(img1)
    print(result.multi_hand_landmarks)
    if result.multi_hand_landmarks:
        for landmarks in result.multi_hand_landmarks:
            mp_drawing.draw_landmarks(img,landmarks,mphands.HAND_CONNECTIONS,
                                      mp_drawing.DrawingSpec(color=(0,255,255),thickness=5,circle_radius=5),
                                      mp_drawing.DrawingSpec(color=(0,0,0),thickness=10))

    cv2.imshow('WEBCAM',img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break 
video.release()
cv2.destroyAllWindows()
```
